### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
users = pd.read_csv('dataset.csv')
users.head()
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby('occupation').mean()[['age']]
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
pd.pivot_table(users, index='occupation', columns = 'gender', values = 'user_id', aggfunc = 'count',
fill_value = 0)
dfMale = pd.pivot_table(users, index='occupation', columns = 'gender', values = 'user_id', aggfunc = 'count',
fill_value = 0)
dfMale['Male ratio per occupation'] = (dfMale['M'] / (dfMale['M'] + dfMale['F']))
sortMale = dfMale.sort_values(by='Male ratio per occupation', ascending = False)
sortMale
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
minMax = users.groupby('occupation').agg({'age':['min','max']})
minMax
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
pd.pivot_table(users, index='occupation', columns = 'gender', values = 'age', aggfunc = 'mean',
fill_value = 0)
```

### Step 8.  For each occupation present the percentage of women and men

```python
pd.pivot_table(users, index='occupation', columns = 'gender', values = 'user_id', aggfunc = 'count',
fill_value = 0)
perc = pd.pivot_table(users, index='occupation', columns = 'gender', values = 'user_id', aggfunc = 'count',
fill_value = 0)
perc['Men (%)'] = (perc['M'] / (perc['M'] + perc['F']) * 100)
perc['Women (%)'] = (perc['F'] / (perc['M'] + perc['F']) * 100)
perc
```
